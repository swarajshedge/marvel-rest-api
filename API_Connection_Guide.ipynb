## Guide to connect to the FLASK API (with illustrations for authentication and various methods)
<br>

<span style='color:Green'> GET-1  </span> <br>
`response=requests.get("http://127.0.0.1:5000/characters").json()`<br>

<span style='color:Green'> GET-2  </span> <br>
`params={"Character Name":["Drax","Harsha","Mantis"]}` <br>
`response=requests.get("http://127.0.0.1:5000/characters",params=params).json()`<br>

<span style='color:Green'> GET-3  </span> <br>
`params={"Character ID":["1011012","1011012","1011077"]}` <br>
`response=requests.get("http://127.0.0.1:5000/characters",params=params).json()`<br>
<br>
<br>
<span style='color:Green'> Signing up  </span> <br>
`params={"email":"99harshavardhan@gmail.com","password":"WinterIsComing"}` <br>
`response=requests.post("http://127.0.0.1:5000/signup",params=params).json()`<br>

<span style='color:Green'> Logging in  </span> <br>
`params={"email":"99harshavardhan@gmail.com","password":"WinterIsComing"}`  <br>
`response=requests.get("http://127.0.0.1:5000/login",params=params).json()`<br>
`access_token=response["token"]`<br>

<span style='color:RED'> INCLUDE THIS access_token AS AN HEADER TO AUTHENTICATE THE POST, DELETE & PUT METHODS </span> <br>
<br>
<br>
<span style='color:Green'> POST-1  </span> <br>
`params={"Character Name":"Thor", "Character ID":"1009999", "Total Available Events":12,
           "Total Available Series":21, "Total Available Comics": 34,
           "Price of the Most Expensive Comic":13.33}` <br>
`headers={'Authorization': f'Bearer {access_token}'}`  <br>
`response=requests.post("http://127.0.0.1:5000/characters",params=params,headers=headers).json()`<br>

<span style='color:Green'> POST-2  </span> <br>
`params={"Character ID":"1009664"}` <br>
`headers={'Authorization': f'Bearer {access_token}'}` <br>
`response=requests.post("http://127.0.0.1:5000/characters",params=params,headers=headers).json()`<br>
<br>
<br>
<span style='color:Green'> DELETE-1  </span> <br>

`response=requests.delete("http://127.0.0.1:5000/characters",params=params,headers=headers).json()`<br>

<span style='color:Green'> DELETE-2  </span> <br>

<br>
<br>
